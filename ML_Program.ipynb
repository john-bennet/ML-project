```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# 1. Load supervised dataset
data = load_iris()
X = data.data
y = data.target

# 2. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Bagging model (using Decision Tree)
bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=10,
    random_state=42
)
bagging.fit(X_train, y_train)

# 4. Boosting model (AdaBoost)
boosting = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=50,
    random_state=42
)
boosting.fit(X_train, y_train)

# 5. Predictions
bag_pred = bagging.predict(X_test)
boost_pred = boosting.predict(X_test)

# 6. Performance
print("---- BAGGING PERFORMANCE ----")
print("Accuracy:", accuracy_score(y_test, bag_pred))
print(classification_report(y_test, bag_pred))

print("---- BOOSTING PERFORMANCE ----")
print("Accuracy:", accuracy_score(y_test, boost_pred))
print(classification_report(y_test, boost_pred))
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# 1. Load data
data = load_iris()
X = data.data

# 2. K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# 3. Output results
print("Cluster Labels Assigned by K-Means:")
print(kmeans.labels_)

print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```
